```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/spotify_millsongdata.csv', engine='python', on_bad_lines='skip')
df.head()
```

```python
df.drop('link',axis=1,inplace=True)
```

```python
from collections import Counter
count = Counter()
for text in df['text'].dropna():
    words = text.split()
    count.update(words)
frequent_number = pd.DataFrame(count.most_common(30), columns=['word', 'count'])
print(frequent_number)
```

```python
def splitter(text):
    if pd.isna(text):
        return pd.Series([None,None,None,None])
    length=len(text.split())
    part_length=length//4
    words=text.split()
    part1=' '.join(words[:part_length])
    part2=' '.join(words[part_length:2*part_length])
    part3=' '.join(words[2*part_length:3*part_length])
    part4=' '.join(words[3*part_length:4*part_length])
    return pd.Series([part1,part2,part3,part4])
df[['part1','part2','part3','part4']]=df['text'].apply(splitter)
df.head()
```

```python
concat=df.set_index(df.artist + ' - ' + df.song)[['part1','part2','part3','part4']]
concat
```

```python
from sentence_transformers import SentenceTransformer
model=SentenceTransformer('sentence-transformers/all-mpnet-base-v2')
for c in concat.columns:
    concat[c]=concat[c].astype(str)
    embds=model.encode(concat[c].tolist(),show_progress_bar=True,device='cuda')
    concat[c]=list(embds)
```

```python
embd_conc=pd.concat([pd.DataFrame(concat[c].values.tolist()) for c in concat.columns],axis=1).set_index(concat.index)
```

```python
embd_conc= pd.read_pickle('/content/spotify_millsongdata_all_embeddings_parted.pkl')
```

```python
embd_con = embd_conc[~embd_conc.index.duplicated(keep='first')]
```

```python
embd_con
```

```python
from sklearn.metrics.pairwise import cosine_similarity
csm=cosine_similarity(embd_con)
similarity_data=pd.DataFrame(csm,index=embd_con.index,columns=embd_con.index)
similarity_data
```

```python
def top_similar_songs(song,top_n=5):
    if song not in similarity_data.index:
        return f"Song '{song}' not found in the dataset."
    similar_songs = similarity_data[song].sort_values(ascending=False).head(top_n + 1).index.tolist()
    similar_songs.remove(song)  # remove the song itself
    return similar_songs
top_similar_songs('Adele - All I Ask', top_n=5)

```
